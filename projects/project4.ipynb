```python
#learning linear regression model from scratch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([800,1000,1200,1400,1600]) #yo feature ho
y_train=np.array([150,200,250,300,350]) #yo target ho
```

```python
def predict(x,w,b):
    return w*x+b
def compute_cost(x,y,w,b):
    m=x.shape[0] #m vaneko number of training example
    total_cost=0
    for i in range(m):
        f_wb=w*x[i]*b
        total_cost+=(f_wb-y[i])**2
    return (total_cost)/(2*m)
```

```python
def compute_gradient(x,y,w,b):
    m=x.shape[0]
    dj_dw=0
    dj_db=0
    for i in range(m):
        f_wb=w*x[i]+b
        dj_dw+=(f_wb-y[i])*x[i]
        dj_db+=(f_wb-y[i])
    dj_dw/=m
    dj_db/=m
    return dj_dw,dj_db
```

```python
def gradient_descent(x,y,w,b,alpha,max_iters):
    for i in range(max_iters):
        dj_dw,dj_db=compute_gradient(x,y,w,b)
        w=w-alpha*dj_dw
        b=b-alpha*dj_db
        if i%100 ==0:
            cost=compute_cost(x,y,w,b)
            print(f"Iteration{i}:Cost {cost},w {w},b {b}")
    return w,b
```

```python
w_init=0
b_init=0
alpha=0.0000001
iterations=1000

w_final,b_final=gradient_descent(x_train,y_train,w_init,b_init,alpha, iterations)

print(f"Trained Parameters : w={w_final}, b={b_final}")
```

```python
x_test=np.array([1800,2000])
predictions=predict(x_test,w_final,b_final)

print(f"Predictions are - {predictions}")
```

```python
plt.scatter(x_train,y_train, label="Training Data")
plt.plot(x_train,predict(x_train,w_final,b_final),color='red',label='Regression Line')
plt.xlabel("House size in sq ft")
plt.ylabel("Price in $1000s")
plt.title("Linear regression fit")
plt.legend()
plt.show()
```
